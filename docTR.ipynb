```python
from doctr.io import DocumentFile
from doctr.models import ocr_predictor

# Load a pre-trained model
model = ocr_predictor(det_arch='db_resnet50', reco_arch='crnn_vgg16_bn', pretrained=True)
```

```python
# Load a PDF or image
doc = DocumentFile.from_pdf("./inv1.pdf")
# For images
# doc = DocumentFile.from_images("path/to/your/image.jpg")

# Perform OCR on the document
result = model(doc)

# result.show(doc)
```

```python
print(result)
```

```python
# Load a PDF or image
doc2 = DocumentFile.from_pdf("./JS.pdf")
# For images
# doc = DocumentFile.from_images("path/to/your/image.jpg")

# Perform OCR on the document
result = model(doc2)

print(result)
```

```python
def process_doctr_output(doctr_output):
    # Assuming doctr_output is the string representation you provided
    # In practice, you'd likely be working with the actual Python objects

    # For demonstration, we'll just print out the text content
    for page in doctr_output.pages:
        for block in page.blocks:
            for line in block.lines:
                line_text = " ".join([word.value for word in line.words])
                print(line_text)
```

```python
process_doctr_output(result)
```
